```r
# Exercise 3

# 3.1

setwd("/Users/lucaagosti/AgostiLuca_rlab01")
library(tibble)

Data.Coal=read.csv("coal-production-by-country.csv")
Coal.Production <- data.frame(Data.Coal[c(1:11528),])
names(Coal.Production) <- c("Entity", "Code", "Year", "Coal production (TWh)")
as_tibble(Coal.Production)
```

```r
# 3.2

barplot(table(Coal.Production[,1]))
```

```r
# 3.3

#I create a new tibble with only the years past 1970

Coal.Prod1970 <- Coal.Production[Coal.Production[,"Year"]>1969,]
table1 <- table(Coal.Prod1970[,1])
#Coal.Prod1970

#Creating a vectors of 0s that I will fill up in the cycle

Integrated.Production <- rep(12000, 0)

#Index starts from 0 beacuse I need it for the table, not a mistake

for (j in 0:(length(table1)-1))
    {
    Integrated.Production[j+1] <- sum(Coal.Prod1970[sum(table1[0:j]):(sum(table1[0:j])+table1[j+1]),4])
}

#I give a name to every value of the production after 1970

names(Integrated.Production) <- unique(Coal.Production[,"Entity"])
cat("Total coal production from 1970:\n")
Integrated.Production

#Now I can calculate the top 5 producers

Max.Production1970 <- Integrated.Production[order(Integrated.Production, decreasing=TRUE)]
cat("The largest coal producers from 1970 are:\n")
Max.Production1970[1:5]
```

```r
# 3.4

Top5.Producers <- c("World", "Asia Pacific", "Asia and Oceania", "China", "OECD")

#For every Entity in the Top 5 I look for the production values in the "Coal.Prod1970" tibble and plot it in respect to time

for (i in 1:5)
    {
    Top5.Coal <- Coal.Prod1970[Coal.Prod1970[,"Entity"]==Top5.Producers[i],]
    plot(Top5.Coal[,"Year"],Top5.Coal[,"Coal production (TWh)"])
    cat("Coal production trend from 1970 of:",Top5.Producers[i])
}
```

```r
# 3.5

World.cumsum <- cumsum(Coal.Prod1970[Coal.Prod1970[,"Entity"]=="World",4])
plot(Coal.Prod1970[Coal.Prod1970[,"Entity"]=="World",3],World.cumsum)
cat("Cumulative world's coal production from 1981:")
```
